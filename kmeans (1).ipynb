```python
#clusters = 5
#iterations = 5
number_of_centroids = clusters
centroid_array = []
mean_array = []
cluster_array = [clusters]

#populate centroid array with random centroids
for i in range(0,clusters):
    rand_centroid= randint(0, 1499)
    centroid_array.append(data_array[rand_centroid])

data_rows = len(data_array)
#mean array is size of centroid array, number of centroid tuples
mean_array_rows = len(mean_array)


'''
centroid[0] = updatedCentroidVal1
centroid[1] = updatedCentroidVal2
centroid[2] = updatedCentroidVal3
'''
centroid_array = np.array(centroid_array)

#we want 1500rows by number of centroids
mean_array_shape=(clusters)
mean_array = np.zeros(mean_array_shape)

#initialize cluster list
#clust{ 0: [x, y], [x, y], 1: [x, y], [x, y], 2: [x, y], [x, y] }
clust = {}
for i in range(0,clusters):
    clust[i]=[]
    
def assign_data_to_centroid(data_array, centroid_array, mean_array, clust):
    '''
    traverse data_array
    compute distance between each data_point and centroid
    assign data_points to new clusters
    lines 55:
    this is where we assign data_points to new clusters
    get minimum distance from mean_array
    store that point in cluster_array
    cluster_array[index(min(mean_array))] = data_point
    return index -- each index represents centroid/cluster of smallest value, store it in cluster_array
    '''
    for i in range(0,rows):
        data_point = data_array[i]
        #print(data_point)
        #compute distance from each centroid, at each dataPoint
        for j in range(number_of_centroids):
            centroid = centroid_array[j]
            distance = np.sqrt(np.sum(np.square(data_point - centroid)))
            mean_array[j] = distance
            #print(mean_array)
        index = mean_array.argmin()
        clust[index].append(data_point)
        #initialize array again so we can fill with new means
        mean_array.fill(0)
    #print(clust)
    return clust

def compute_centroids(clust,centroid_array):
    '''
    recompute cluster center/centroids
    for each index in cluster array, get average of all points at that index,
    append this value to centroid 
    '''
    for i in range(0,clusters):
        
        previous_centroid = centroid_array[i]
        #print('PREV_centroid',centroid_array[i])
        #GET NEW CENTROID
        new_centroid = np.mean(clust[i], axis = 0)
        #print('previous_centroid: ',previous_centroid)
        print('new_centroid',i,new_centroid)
        centroid_array[i]= new_centroid
        #COMPUTE MEANS SQ. ERROR
        mse = ((new_centroid - previous_centroid)**2).mean(axis=0)
        #mse = (np.square(new_centroid - new_centroid)).nanmean(axis=0)
        #mse = mean_squared_error(previous_centroid, new_centroid)
    print('Squared Error: ',mse)
    #print(centroid_array)
    #print('-------------')
    return centroid_array

def scatter_plot_data(clusters, data_array):
    #"*zip" unzips "*data_array"
    #(meaning it uncouples the tuple values inside and returns two temp arrays, x and y with the separated values)
    #"*data_array" unpacks the entries inside data_array into discrete segments for use in plot.scatter
    plot.scatter(*zip(*data_array), 2)
    plot.show()
    plot.clf()

def scatter_plot_data_color(clusters, data_array, centroid_array, mean_array, clust):
    #"*zip" unzips "*data_array"
    #returns two temp arrays, x and y with the diff values)
    #"*data_array" unpacks the entries inside data_array into discrete segments for use in plot.scatter
    scatter_plot = plot.scatter(*zip(*data_array), 2, color="b")
    scatter_plot = plot.scatter(*zip(*centroid_array), 60, color="r")
    scatter_plot = plot.scatter(*zip(*clust[0]), 2, color="b")
    plot.show()
    plot.clf()
    return scatter_plot

def scatter_plot(clusters, data_array, centroid_array, mean_array, clust):
    colors = 10 * ["b", "m", "g", "y", "c", "k"]
    for i in range(0,clusters):
        color = colors[i]
        cluster_array = clust[i]
        centroid_array= centroid_array
        plot.scatter(*zip(*cluster_array), 3, color=color)
        plot.scatter(*zip(*centroid_array), 10, color="r")
    plot.show()
    
#Main For Loop

for i in range(0,iterations):
    clust = assign_data_to_centroid(data_array, centroid_array, mean_array, clust)
    centroid_array = compute_centroids(clust,centroid_array)
    #plot1 = scatter_plot_data_color(clusters, data_array, centroid_array, mean_array, clust)
    plot2 = scatter_plot(clusters, data_array, centroid_array, mean_array, clust)
    print('-----Iteration-----',i)
```
